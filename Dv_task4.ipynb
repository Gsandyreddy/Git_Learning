```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('restaurant_dataset.csv')
print(df)
# Scatterplot Matrix
sns.pairplot(df, hue='Restaurant')
plt.title('Scatterplot Matrix')
plt.show()
```

```python

# Parallel Coordinates
from pandas.plotting import parallel_coordinates
plt.figure(figsize=(10, 6))
parallel_coordinates(df, 'Restaurant', colormap='viridis')
plt.title('Parallel Coordinates')
plt.show()
```

```python
# Line Graph
plt.figure(figsize=(10, 6))
for restaurant in df['Restaurant'].unique():
    data = df[df['Restaurant'] == restaurant]
    plt.plot(data.columns[1:], data.iloc[:, 1:].values.flatten(), label=restaurant)
```

```python
plt.title('Line Graph')
plt.xlabel('Attributes')
plt.ylabel('Performance Ratings')
plt.legend()
plt.show()
```

```python
# Stacked Bar Chart
df.set_index('Restaurant').T.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Stacked Bar Chart')
plt.xlabel('Attributes')
plt.ylabel('Performance Ratings')
plt.legend(title='Restaurant')
plt.show()
```

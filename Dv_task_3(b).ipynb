```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load the Credit Card Default dataset
df=pd.read_csv("/content/airbnb_dataset.csv")
print(df)
# Categorical vs. Categorical
# Stacked Bar Chart
stacked_bar_airbnb = df.groupby(['Location', 'Number_of_Bedrooms']).size().unstack()
stacked_bar_airbnb.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Chart: Location vs. Number of Bedrooms')
plt.show()
```

```python
# Grouped Bar Chart
grouped_bar_airbnb = df.groupby(['Location', 'Number_of_Bedrooms']).size().unstack()
grouped_bar_airbnb.plot(kind='bar', figsize=(10, 6))
plt.title('Grouped Bar Chart: Location vs. Number of Bedrooms')
plt.show()
```

```python

# Segmented Bar Chart
fig_airbnb = px.histogram(df, x="Location", color="Number_of_Bedrooms", barmode="group")
fig_airbnb.update_layout(title_text='Segmented Bar Chart: Location vs. Number of Bedrooms')
fig_airbnb.show()
```

```python
# Continuous vs. Continuous
# Scatterplot Fit Lines
sns.lmplot(x='Number_of_Bedrooms', y='Price', data=df, height=6, aspect=1.5)
plt.title('Scatterplot Fit Lines: Number of Bedrooms vs. Price')
plt.show()
```

```python
# Categorical vs. Continuous
# Bar Chart (Summary Statistics)
bar_chart_airbnb = sns.barplot(x='Location', y='Price', data=df, ci='sd')
plt.title('Bar Chart: Location vs. Price')
plt.show()
```

```python
# Grouped Kernel Density Plots
sns.FacetGrid(df, col="Location", height=4, col_wrap=3).map(sns.kdeplot, "Price")
plt.suptitle('Grouped Kernel Density Plots: Location vs. Price', y=1.02)
plt.show()
```

```python

# Box Plots
sns.boxplot(x='Location', y='Price', data=df)
plt.title('Box Plot: Location vs. Price')
plt.show()
```

```python

# Violin Plots
sns.violinplot(x='Location', y='Price', data=df)
plt.title('Violin Plot: Location vs. Price')
plt.show()
```

```python
# Ridgeline Plots
sns.displot(df, x="Price", hue="Location", kind="kde", fill=True)
plt.title('Ridgeline Plot: Location vs. Price')
plt.show()
```

```python
# Beeswarm Plots
sns.swarmplot(x='Location', y='Price', data=df)
plt.title('Beeswarm Plot: Location vs. Price')
plt.show()
```

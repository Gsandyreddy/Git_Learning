```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load the Credit Card Default dataset
df=pd.read_csv("/content/credit_card_dataset.csv")
print(df)
# Categorical vs. Categorical
# Stacked Bar Chart
stacked_bar = df.groupby(['Education', 'Default_payment']).size().unstack()
stacked_bar.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Chart: Education vs. Default')
plt.show()
```

```python
# Grouped Bar Chart
grouped_bar = df.groupby(['Education', 'Default_payment']).size().unstack()
grouped_bar.plot(kind='bar', figsize=(10, 6))
plt.title('Grouped Bar Chart: Education vs. Default')
plt.show()
```

```python
# Segmented Bar Chart
fig = px.histogram(df, x="Education", color="Default_payment", barmode="group")
fig.update_layout(title_text='Segmented Bar Chart: Education vs. Default')
fig.show()
```

```python
# Continuous vs. Continuous
# Scatterplot Fit Lines
sns.lmplot(x='Age', y='Limit_bal', data=df, height=6, aspect=1.5)
plt.title('Scatterplot Fit Lines: Age vs. Limit Balance')
plt.show()
```

```python
# Categorical vs. Continuous
# Bar Chart (Summary Statistics)
bar_chart = sns.barplot(x='Education', y='Limit_bal', data=df, ci='sd')
plt.title('Bar Chart: Education vs. Limit Balance')
plt.show()
```

```python
# Grouped Kernel Density Plots
sns.FacetGrid(df, col="Education", height=4, col_wrap=3).map(sns.kdeplot, "Limit_bal")
plt.suptitle('Grouped Kernel Density Plots: Education vs. Limit Balance', y=1.02)
plt.show()
```

```python

# Box Plots
sns.boxplot(x='Education', y='Limit_bal', data=df)
plt.title('Box Plot: Education vs. Limit Balance')
plt.show()
```

```python

# Violin Plots
sns.violinplot(x='Education', y='Limit_bal', data=df)
plt.title('Violin Plot: Education vs. Limit Balance')
plt.show()
```

```python
# Ridgeline Plots
sns.displot(df, x="Limit_bal", hue="Education", kind="kde", fill=True)
plt.title('Ridgeline Plot: Education vs. Limit Balance')
plt.show()
```

```python

# Beeswarm Plots
sns.swarmplot(x='Education', y='Limit_bal', data=df)
plt.title('Beeswarm Plot: Education vs. Limit Balance')
plt.show()
```
